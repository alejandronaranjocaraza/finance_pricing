```python
import sys
sys.path.append('../build')
import pricer
import matplotlib.pyplot as plt
```

```python
vanillaOption = pricer.makeVanillaOption(2,3)
```

```python
stock = pricer.makeStock(10,0.05)
```

```python
bsm = pricer.makeBSM(vanillaOption,stock, 0.1, 0.5)
```

```python
x = [0.5*i for i in range (100)]
y = []
for targetPrice in x:
    y.append(bsm.impliedVol(targetPrice))
```

```python
plt.plot(x,y)
```
